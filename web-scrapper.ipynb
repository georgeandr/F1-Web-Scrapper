### This cell creates a dataframe with the driver's Position, Name, Nationality, Car manufacturer, Points scored, and the Year

```python
drivers_df = None

for item in list_of_urls:
    url1 = 'https://www.formula1.com'
    url = url1 + item
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'lxml')
    standings_table = soup.find('table', class_ = 'resultsarchive-table')
    if drivers_df is None:
        headers = []
        for i in standings_table.find_all('th'):
            title = i.text
            headers.append(title)
        drivers_df = pd.DataFrame(columns = headers)
        drivers_df['Year'] = ''
    for tr in standings_table.find_all('tr')[1:]:
        row_data = tr.find_all('td')
        row = [i.text for i in row_data]
        year_data = soup.find('h1', class_ = 'ResultsArchiveTitle')
        year = [i.text for i in year_data]
        row.append(year)
        length = len(drivers_df)
        drivers_df.loc[length] = row         
```
