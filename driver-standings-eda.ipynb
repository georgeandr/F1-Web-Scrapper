```python
import numpy as np
import pandas as pd
import seaborn as sns
import pycountry
from matplotlib import pyplot as plt
import plotly
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as pyo
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
standings_df = pd.read_csv('driver_stangings.csv', delimiter = ';')
standings_df
```

```python
standings_df['Year'] = standings_df['Year'].str.rstrip(' Driver Standings')
standings_df.head(10)
```

```python
def annually_drivers_points(df, year):
    year_df = df.loc[df['Year'] == year]
    f, ax = plt.subplots(figsize=(10, 5))
    ax = sns.barplot(data = year_df, x = 'Driver', y = 'PTS', palette = 'rocket')
    ax = plt.xticks(rotation = 70)
    ax = plt.ylabel('Points Scored in %s' %year)
    ax = plt.xlabel('Driver\'s Name')
    plt.title("Drivers Points of %s" %year)
```

```python
annually_drivers_points(standings_df, '2021')
```

```python
def driver_points(df, name):
    df_name = df.loc[df['Driver'] == name]
    f, ax = plt.subplots(figsize=(10, 5))
    ax = sns.barplot(data = df_name, x = 'Year', y = 'PTS', palette = 'viridis')
    ax = plt.xticks(rotation = 70)
    ax = plt.ylabel('Points Scored')
    ax = plt.xlabel('Driver\'s Name')
    plt.title("Points of %s" %name + "scored from the beggining of his career")
```

```python
driver_points(standings_df, 'Lewis Hamilton HAM')
```
